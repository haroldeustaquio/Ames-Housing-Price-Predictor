### Important Features

| Variable         | Justificación                                                                                            |
| ---------------- | -------------------------------------------------------------------------------------------------------- |
| `Neighborhood`   | La ubicación es **el factor más importante** en bienes raíces. Hay zonas premium y otras más accesibles. |
| `Overall Qual`   | Refleja la calidad general del material y acabado. Directamente ligado al valor de la casa.              |
| `Gr Liv Area`    | Área habitable sobre el suelo. A mayor área útil, mayor valor.                                           |
| `Garage Cars`    | Tener garaje (y su capacidad) aumenta significativamente el valor.                                       |
| `Garage Area`    | Área del garaje influye tanto como el número de autos.                                                   |
| `Total Bsmt SF`  | Área de sótano también es valorizada, especialmente si está terminado.                                   |
| `Kitchen Qual`   | La cocina es uno de los ambientes más costosos y más observados en una compra.                           |
| `Year Built`     | Año de construcción indica si la casa es moderna o antigua.                                              |
| `Year Remod/Add` | Remodelaciones recientes aumentan el valor.                                                              |
| `Full Bath`      | Los baños completos impactan directamente el valor.                                                      |
| `Fireplaces`     | Tener chimenea mejora el atractivo y valor en climas fríos.                                              |
| `Exter Qual`     | Calidad del exterior es esencial para valoración estética y estructural.                                 |
| `1st Flr SF`     | Área del primer piso suele contener ambientes importantes (sala, cocina).                                |
| `Garage Type`    | Tipo de garaje (adjunto, independiente) influye en funcionalidad y valorización.                         |
| `Mas Vnr Type`   | Revestimientos externos de piedra/ladrillo aportan valor visual y constructivo.                          |
| `Mas Vnr Area`   | Más revestimiento suele estar asociado con casas de mayor costo.                                         |
| `Foundation`     | Tipo de cimentación afecta durabilidad y valor de la construcción.                                       |
| `Lot Area`       | Terreno grande es valioso, especialmente en zonas urbanas.                                               |
| `Central Air`    | Aire acondicionado central es una comodidad muy buscada.                                                 |
| `House Style`    | Algunas configuraciones son más costosas (e.g., 2 pisos vs ranchera).                                    |
| `TotRms AbvGrd`  | Número total de habitaciones habitables es una métrica base en valuación.                                |
| `Garage Finish`  | Tener el garaje terminado aporta valor adicional.                                                        |


### Intermediate Impact

| Variable           | Justificación                                                                            |
| ------------------ | ---------------------------------------------------------------------------------------- |
| `MS Zoning`        | Restricciones legales pueden influir en el valor futuro o uso del inmueble.              |
| `Condition 1/2`    | Proximidad a ferrovías, calles principales, ríos, etc., puede elevar o reducir el valor. |
| `Bldg Type`        | Si es adosada, duplex, etc., afecta el valor comparado a una unifamiliar.                |
| `Exterior 1st/2nd` | Materiales exteriores tienen impacto visual y de durabilidad.                            |
| `Exter Cond`       | Condición externa influye en el mantenimiento requerido.                                 |
| `Bsmt Qual`        | Altura del sótano puede definir si es usable como ambiente.                              |
| `BsmtFin Type 1/2` | Si el sótano está acondicionado como sala/habitación mejora su valor.                    |
| `BsmtFin SF 1/2`   | Superficie acondicionada en sótano agrega espacio útil.                                  |
| `Bsmt Exposure`    | Ventanas o salida al exterior mejora mucho la valoración del sótano.                     |
| `Heating QC`       | Calidad del sistema de calefacción mejora confort (en climas fríos).                     |
| `Electrical`       | Un sistema eléctrico moderno es mejor percibido por compradores.                         |
| `Kitchen AbvGr`    | Tener más de una cocina puede importar en propiedades grandes o multifamiliares.         |
| `Fireplace Qu`     | Calidad de chimenea puede importar si hay una.                                           |
| `Garage Yr Blt`    | Un garaje más moderno puede reflejar mejoras posteriores.                                |
| `Garage Qual/Cond` | Puede afectar si hay problemas de estructura, humedad, etc.                              |
| `Paved Drive`      | Entrada pavimentada es un plus estético y funcional.                                     |
| `Wood Deck SF`     | Espacios exteriores de calidad mejoran el valor.                                         |
| `Open Porch SF`    | Lo mismo que el deck: atractivo en climas templados.                                     |
| `Screen Porch`     | Valor variable, depende de clima y preferencias.                                         |
| `Fence`            | Puede importar en familias con niños o mascotas.                                         |
| `Yr Sold`          | Año de venta importa si el mercado está en crecimiento o crisis.                         |
| `Sale Condition`   | Algunas condiciones (venta forzada, etc.) bajan el precio.                               |
| `Lot Frontage`     | Frente del terreno puede importar en zonas urbanas (fachada amplia).                     |
| `Roof Style`       | Estética y funcionalidad influyen en mantenimiento.                                      |
| `Functional`       | Grado de funcionalidad de la casa puede penalizar si es muy bajo.                        |
| `2nd Flr SF`       | Piso superior agrega valor, pero no tanto como planta baja.                              |
| `Half Bath`        | Tiene menor peso que baño completo, pero suma.                                           |
| `Bedroom AbvGr`    | Aunque importa, se valora más el tamaño/calidad que la cantidad.                         |
| `Screen Porch`     | Valor agregado si el clima lo permite, pero no universal.                                |


### Don't have impact

| Variable          | Justificación                                                                           |
| ----------------- | --------------------------------------------------------------------------------------- |
| `Order`           | Solo es un índice interno.                                                              |
| `PID`             | ID parcelario, no dice nada del inmueble.                                               |
| `Alley`           | La mayoría de las casas no tienen alley (mucha ausencia), y tenerlo no aporta mucho.    |
| `Low Qual Fin SF` | Acabado de baja calidad no añade valor, incluso puede restar.                           |
| `Enclosed Porch`  | Porches cerrados no suelen valorarse tanto como los abiertos o funcionales.             |
| `3Ssn Porch`      | Uso limitado, no siempre valorado por compradores.                                      |
| `Pool Area`       | En Ames (Iowa), las piscinas no se valoran mucho por clima; además son poco frecuentes. |
| `Pool QC`         | Pocas casas con piscina, no es representativo.                                          |
| `Misc Feature`    | Demasiado variado y poco frecuente para ser relevante.                                  |
| `Misc Val`        | Mismo caso: bajo impacto y muy disperso.                                                |
| `Mo Sold`         | El mes no suele influir significativamente en el precio (puede en volumen de ventas).   |
| `Utilities`       | Casi todas las casas tienen todos los servicios → nula variabilidad.                    |
| `Street`          | La mayoría tiene calles pavimentadas, no discrimina bien.                               |
| `Heating`         | Muy poca variación, no influye mucho en el precio por sí sola.                          |
| `Land Slope`      | Pendiente del terreno rara vez es crítica en zonas planas como Ames.                    |
| `Land Contour`    | Similar al anterior; no suele ser un factor de valoración fuerte.                       |
| `MS SubClass`     | Aunque codifica el tipo de vivienda, es mejor usar `Bldg Type` y `House Style`.         |
